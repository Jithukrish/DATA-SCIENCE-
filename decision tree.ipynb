```python
import pandas as pd
# from sklearn.tree import DecisionTreeClassifire as Dt
import matplotlib.pyplot as plt
read = pd.read_csv('iris.csv')

x=read.iloc[:,:4].values

y=read.iloc[:,4:].values
```

```python
#spliting
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#prediction
from sklearn.tree import DecisionTreeClassifier as Dt
from sklearn import tree
import matplotlib.pyplot as plt

c= Dt(criterion='entropy')  

c.fit(x_train,y_train)
plt.figure(figsize=(20,20))
tree.plot_tree(c)
plt.show()
p=c.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
result=confusion_matrix(y_test,p)
print("confusion_matrix:\n ",result)


print("accuracy_score:  ",accuracy_score(y_test,p))
print("classification_report:  ",classification_report(y_test,p))
```
