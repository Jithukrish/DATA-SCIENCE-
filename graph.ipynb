```python
import numpy as n
import matplotlib.pyplot as plt
x=n.array([1,2,6,8])
y=n.array([1,2,6,8])
plt.plot(x,y,ms=15,marker='*')
plt.title('graph')
plt.plot(x,y)
plt.show()
```

```python
import numpy as n
import matplotlib.pyplot as plt
x=n.array([2,4,7,8])
y=[0,0,0.2,0]
d=["apple","orange","banana","mango"]
plt.pie(x,labels=d,explode=y,shadow=True)
plt.legend(loc='upper right')
plt.show()
```
