```python
#rotate
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('flower.jpg')
b=np.array(a)
c=ndimage.rotate(b,270,cval=200)
d=Image.fromarray(c)
d.save('flower.jpg')
plt.imshow(d)
```

```python
#scaling
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('flower.jpg')
b=np.array(a)
#c=ndimage.rotate(b,90,cval=200)

c=ndimage.zoom(b,(0.25,0.5,1))
d=Image.fromarray(c)
d.save('flower1.jpg')
plt.imshow(d)
```
